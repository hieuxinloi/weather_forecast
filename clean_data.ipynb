# Clean data

```python
import pandas as pd
```

```python
data = pd.read_csv('data/origin/weatherQN_2021_2025.csv')
data.head(10)
```

```python
data.info()
```

## Drop Null columns

```python
data.drop(columns=['snow', 'wpgt', 'tsun'], inplace=True)
```

```python
data.info()
```

```python
data.head(10)
```

## Fill missing values

### Fill coco with nearest values because it is the weather condition.

```python
data['coco'].bfill(inplace=True)
```

```python
data.head(10)
```

### Fill prcp with median values

```python
data['prcp'] = data['prcp'].fillna(data['prcp'].mean())
```

```python
data.head(10)
```

## Save cleaned data

```python
data.to_csv('data/cleaned/weatherQN_2021_2025_cleaned.csv', index=False)
```
