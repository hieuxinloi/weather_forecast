```python
# Get the nearby weather stations
stations = Stations()
stations = stations.nearby(13.7667, 109.2167)
stations = stations.fetch(1)

stations
```

```python
%%time

# Get weather data with wmo
start = datetime(2021, 1, 1)
end = datetime(2025, 1, 19, 23, 59)

data = Hourly('48870', start, end)
data = data.fetch()
```
